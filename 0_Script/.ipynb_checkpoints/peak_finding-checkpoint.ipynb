```python
import os
import pandas as pd
from datetime import datetime
import glob
from preprocessing_compiled import read_compiled_files, create_monitors_dict
from find_peaks import catch_peaks, find_non_overlapping_peaks_combined, process_catch_peaks_output
from find_minimum import catch_mins, find_non_overlapping_mins_combined, process_catch_mins_output
from visualize_with_peaks import plot_df
from extract_value_at_peak import extract_monitors_value_at_peak
```

### Customization

```python
#method = "valley"
method = "peak"
```

```python
co2_prominence = 1400
tvoc_prominence = 300
pm2p5_prominence = 15
temp_prominence = 3
humidity_prominence = 15
```

```python
window_hours = 20
```

```python
min_co2 = 300
max_co2 = 2500

max_pm2p5 = 40

max_tvoc = 1300

min_temp = 23
max_temp = 30

min_humidity = 35
max_humidity = 80
```

### Preprocessing

```python
folder_path = '../1_Data/1_param_compiled'
params_dict = read_compiled_files(folder_path)
monitors_df_dict = create_monitors_dict(params_dict)
```

### Peak Catching

```python
output_dir = "../2_Visualizations"
os.makedirs(output_dir, exist_ok=True)

if method == "peak":
    peaks_dict = catch_peaks(monitors_df_dict, co2_prominence, tvoc_prominence, pm2p5_prominence, temp_prominence, humidity_prominence)
    final_peaks_dict = process_catch_peaks_output(peaks_dict, window_hours)
    report_fig = {
        param: plot_df(param, data, final_peaks_dict, min_co2, max_co2,max_pm2p5, max_tvoc, min_temp, max_temp, min_humidity, max_humidity) 
        for param, data in params_dict.items()
    }
    peak_dict = extract_monitors_value_at_peak(params_dict, final_peaks_dict)
    
    for param, fig in report_fig.items():
        fig.savefig(f"{output_dir}/{param}_peak.png")
        
    for filename, dataframe in peak_dict.items():
        dataframe.to_csv(f"../1_Data/2_peak/{filename}_peak.csv", index=False)
    
elif method == "valley":
    valleys_dict = catch_mins(monitors_df_dict, co2_prominence, tvoc_prominence, pm2p5_prominence, temp_prominence, humidity_prominence)
    final_valleys_dict = process_catch_mins_output(valleys_dict, window_hours)
    report_fig = {
        param: plot_df(param, data, final_valleys_dict, min_co2, max_co2,max_pm2p5, max_tvoc, min_temp, max_temp, min_humidity, max_humidity) 
        for param, data in params_dict.items()
    }
    valley_dict = extract_monitors_value_at_peak(params_dict, final_valleys_dict)

    for param, fig in report_fig.items():
        fig.savefig(f"{output_dir}/{param}_valley.png")

    for filename, dataframe in valley_dict.items():
        dataframe.to_csv(f"../1_Data/2_peak/{filename}_valley.csv", index=False)
    
```
